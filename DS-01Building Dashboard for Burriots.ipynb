```python
import atoti as tt
session = tt.Session()
# session = tt.Session(config={'user_content_storage': './content', 'port': 9000})
```

```python
#normalization of the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
burrito_vars_norm = df.loc[:, ['Circum', 'Volume', 'Length', 'Mass','Cost']]
bnorms = scaler.fit_transform(burrito_vars_norm)*10
df[['Circum_norm', 'Volume_norm', 'Length_norm', 'Mass_norm','Cost_norm']] = bnorms
# burrito_vars_norm.isnull().sum()
```

```python
df.columns
```

```python
burritos_variables = pd.melt(df.reset_index(), id_vars = ['Location', 'Burrito'], value_vars = ['Circum_norm', 'Volume_norm', 'Length_norm', 'Mass_norm','Cost_norm'])
burritos_variables
```

```python
burrito_table = session.read_pandas(df, table_name = 'burritos')
```

```python
burrito_table.head()
```

```python
cube = session.create_cube(burrito_table)
```

```python
h = cube.hierarchies
l = cube.levels
m = cube.measures
```

```python
m['five'] = 5
m['lenXwrap'] = m['Length.MEAN'] * m['Wrap.MEAN']
```

```python
session.visualize('exploration 1')
```

```python
session.visualize('scatter_plot neighbourhood yelp google')
```

```python
burrito_var_table = session.read_pandas(burritos_variables, table_name='burritos_variables', keys=['Location', 'Burrito', 'variable'])
```

```python
burrito_table.join(burrito_var_table)
```

```python
m['value'] = tt.agg.mean(burrito_var_table["value"])
m['agvalue'] = tt.agg.mean(m['value'], scope = tt.scope.origin(l['Location'], l['Burrito'], l['variable']))
```
